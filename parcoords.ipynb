```python
import pandas as pd
```

```python
import plotly.graph_objs as go
import plotly
from plotly.offline import plot, iplot, init_notebook_mode
init_notebook_mode(connected = True)
import plotly.offline as offline
```

```python
cars = pd.read_csv('C:/Users/Administrator/Desktop/Experiments/cars.csv')
```

```python

trace = go.Parcoords(line={'color':'green'},
    dimensions=list([
        {'values':cars.MPG,'label':'MPG'},
        {'values':cars.Cylinders,'label':'Cylinders'},
        {'values':cars.Horsepower,'label':'Horsepower'},
        {'values':cars.Weight,'label':'Weight'}
    ]),
    labelfont = {'color':'red'},
    tickfont = {'color':'blue'},
    rangefont = {'color':'purple'},              
)
layout = go.Layout(title='Test Title',
                   titlefont={'size':40,'color':'blue'},
   )
fig = go.Figure(data=[trace],layout=layout)
offline.iplot(fig)
```
