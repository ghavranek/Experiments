```python
from sklearn import datasets
import pandas as pd
```

```python
iris = datasets.load_iris()
```

```python
iris = iris.data[:,:]
```

```python
x = pd.DataFrame(iris)
```

```python
from scipy.spatial.distance import mahalanobis
import scipy as sp
```

```python
Sx = x.cov().values
```

```python
Sx = sp.linalg.inv(Sx)
```

```python
mean=x.mean().values
```

```python
def mahalanobisR(X,meanCol,IC):
    m=[]
    for i in range(len(x)):
        m.append(mahalanobis(X.iloc[i,:],meanCol,IC) ** 2)
    return(m)

```

```python
mR = mahalanobisR(x,mean,Sx)
```

```python
from scipy import stats
```

```python
qq1 = stats.probplot(mR,dist='chi2',sparams = 1)
```

```python
qq3 = stats.probplot(mR,dist='chi2',sparams = 3)
```
